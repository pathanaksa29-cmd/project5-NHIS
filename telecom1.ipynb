```python
def check_skewness(col):
    val = col.skew()
    
    if -0.5 <= val <= 0.5:
        return "Symmetrical (Normal-ish)"
    elif 0.5 < val <= 1:
        return "Moderately Positive Skewed"
    elif val > 1:
        return "Highly Positive Skewed"
    elif -1 <= val < -0.5:
        return "Moderately Negative Skewed"
    else:
        return "Highly Negative Skewed"

def check_kurtosis(col):
    val = col.kurt()
    
    if val > 0.5:
        return "Leptokurtic (Heavy Tails / Peaked)"
    elif val < -0.5:
        return "Platykurtic (Light Tails / Flat)"
    else:
        return "Mesokurtic (Normal-like)"
```

```python
# Select Features for Clustering

features = df_engagement[
    [
        "Session_Frequency",
        "Total_Duration_ms",
        "Total_Traffic"
    ]
].copy()

# Handle Missing Values

features = features.fillna(features.mean())

# Normalize Data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

engagement_norm = scaler.fit_transform(features)

# Elbow Method

inertia = []

for k in range(1, 11):

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(engagement_norm)

    inertia.append(model.inertia_)

# Plot Elbow Curve

plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    inertia,
    marker='o',
    linewidth=2
)

plt.title("Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(range(1,11))
plt.grid(True)

plt.show()

# Final KMeans (K = 3)

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_engagement["Engagement_Cluster"] = kmeans.fit_predict(engagement_norm)

# Cluster Summary

cluster_summary = (
    df_engagement
    .groupby("Engagement_Cluster")
    .agg({
        "Session_Frequency": "mean",
        "Total_Duration_ms": "mean",
        "Total_Traffic": "mean"
    })
)

print(cluster_summary)

# First 5 Rows

print(df_engagement.head())
```

```python
def display_metric_statistics(df, column):
    print("=" * 80)
    print(f"{column}")
    print("=" * 80)

    print("\nTop 10 Values")
    display(df.nlargest(10, column)[["MSISDN/Number", column]])

    print("\nBottom 10 Values")
    display(df.nsmallest(10, column)[["MSISDN/Number", column]])

    print("\nMost Frequent Values")
    display(
        df[column]
        .value_counts()
        .head(10)
        .reset_index()
        .rename(columns={"index": column, column: "Frequency"})
    )
```

# Task 4. User Satisfaction Analysis

```python
engagement_kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_engagement["Engagement_Cluster"] = engagement_kmeans.fit_predict(
    engagement_scaled
)

experience_kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_experience["Experience_Cluster"] = experience_kmeans.fit_predict(
    experience_scaled
)
```

```python
worst_experience_cluster = 2

experience_distance = cdist(
    experience_scaled,
    experience_kmeans.cluster_centers_
)

df_experience["Experience_Score"] = experience_distance[
    :,
    worst_experience_cluster
]
```

```python
satisfaction_df = df_engagement.merge(
    df_experience[
        [
            "MSISDN/Number",
            "Experience_Score"
        ]
    ],
    on="MSISDN/Number"
)
```
